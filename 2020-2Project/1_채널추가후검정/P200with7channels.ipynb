```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pprint import pprint
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc
```

```python
EEG_df = pd.read_csv('P200.csv')
EEG_df.head()
```

```python
col_names = EEG_df.columns.values
X = EEG_df[col_names[0:]]
Y = EEG_df[col_names[2:4]]
```

```python
X = X.drop(['subject','trial'], axis=1)
X.head(5)
```

```python
X_condition1=X[X['condition']==1]
X_condition2=X[X['condition']==2]
X_condition3=X[X['condition']==3]
```

```python
X_condition1=X_condition1.drop('condition',axis=1)
X_condition2=X_condition2.drop('condition',axis=1)
X_condition3=X_condition3.drop('condition',axis=1)
```

```python
X_condition1=X_condition1.drop('group',axis=1)
X_condition2=X_condition2.drop('group',axis=1)
X_condition3=X_condition3.drop('group',axis=1)
X_condition1.head(5)
```

```python
Y_condition1=Y[Y['condition']==1]
Y_condition2=Y[Y['condition']==2]
Y_condition3=Y[Y['condition']==3]
```

```python
Y_condition1=Y_condition1.drop('condition',axis=1)
Y_condition2=Y_condition2.drop('condition',axis=1)
Y_condition3=Y_condition3.drop('condition',axis=1)
Y_condition1.head(5)
```

```python
def train_val_test_split(X, Y, test_size=0.3, random_state=123):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
                                                        test_size=test_size, 
                                                        random_state=random_state)
   
    return X_train, X_test, Y_train, Y_test
```

```python
X_condition1_train, X_condition1_test, Y_condition1_train, Y_condition1_test = train_val_test_split(X_condition1, Y_condition1,
                                                                      
                                                                      test_size=0.3,
                                                                      random_state=123)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
rkfold = RepeatedKFold(n_splits=7, n_repeats=3, random_state=42) 
param_grid = {'weights':['uniform', 'distance'],
'n_neighbors': [9, 11, 13, 15, 17]}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=rkfold, return_train_score=True, scoring='accuracy', iid=False)
grid.fit(X_condition1_train, Y_condition1_train)
grid.score(X_condition1_test, Y_condition1_test), grid.best_params_, grid.best_score_
```

```python
rkfold = RepeatedKFold(n_splits=7, n_repeats=3, random_state=42) #tol,intercept_scaling, class_weight, random_state
param_grid = {'n_estimators': [15, 20, 25],
'max_features': ['auto', 'sqrt', 'log2']}
grid = GridSearchCV(RandomForestClassifier(), param_grid, cv=rkfold, return_train_score=True, scoring='accuracy', iid=False)
grid.fit(X_condition1_train, Y_condition1_train)
grid.score(X_condition1_test, Y_condition1_test), grid.best_params_, grid.best_score_
```

```python
rkfold = RepeatedKFold(n_splits=7, n_repeats=3, random_state=42) #tol,intercept_scaling, class_weight, random_state
param_grid = {'priors': [None, [0.5, 0.5], [0.6, 0.4], [0.7, 0.3], [0.8, 0.2], [0.9, 0.1]]}
grid = GridSearchCV(GaussianNB(), param_grid, cv=rkfold, return_train_score=True, scoring='accuracy', iid=False)
grid.fit(X_condition1_train, Y_condition1_train)
grid.score(X_condition1_test, Y_condition1_test), grid.best_params_, grid.best_score_
```

```python
rkfold = RepeatedKFold(n_splits=7, n_repeats=3, random_state=42) #tol,intercept_scaling, class_weight, random_state
param_grid = {'class_weight': [None, 'balanced'],
'max_depth': [3, 4, 5, 6, 7]}
grid = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=rkfold, return_train_score=True, scoring='accuracy', iid=False)
grid.fit(X_condition1_train, Y_condition1_train)
grid.score(X_condition1_test, Y_condition1_test), grid.best_params_, grid.best_score_
```

```python
rkfold = RepeatedKFold(n_splits=7, n_repeats=3, random_state=42) #tol,intercept_scaling, class_weight, random_state
param_grid = {'penalty': ['l1', 'l2'],
'C': [0.1, 1, 10, 1e2, 1e3, 1e4, 1e5, 1e6]}
grid = GridSearchCV(LogisticRegression(), param_grid, cv=rkfold, return_train_score=True, scoring='accuracy', iid=False)
grid.fit(X_condition1_train, Y_condition1_train)
grid.score(X_condition1_test, Y_condition1_test), grid.best_params_, grid.best_score_
```
